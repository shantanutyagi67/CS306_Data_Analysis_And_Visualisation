```python
import seaborn as sns
import csv
import matplotlib.pyplot as plt
import os
import numpy as np
from matplotlib import colors 
from matplotlib.ticker import PercentFormatter 
import pandas as pd
import scipy.stats
%matplotlib inline
```

```python
df = pd.read_csv (r'2008.csv')
df=df[np.isfinite(df['AirTime'])]
df=df[np.isfinite(df['Distance'])]
airtime_data_original=df['AirTime']
distance_data_original=df['Distance']
```

```python
def normalize_data(data_array,mean,sigma):
    ans=[]
    for i,j in data_array.iteritems():
        ans.append((j-mean)/sigma)
    return pd.Series(ans)
```

```python

def find_outlier(data_array):

    Q1 = np.percentile(data_array, 25, interpolation = 'linear')  
    Q2 = np.percentile(data_array, 50, interpolation = 'linear')  
    Q3 = np.percentile(data_array, 75, interpolation = 'linear') 
    iqr=scipy.stats.iqr(data_array)
    outlier=[]
    for x in data_array:
        if ((x> Q3+1.5*iqr) or (x<Q1-1.5*iqr)): 
         outlier.append(x)
    return Q1,Q3,iqr,outlier
# df_filtered=df_filtered[df['TMAX']>=Q1-1.5*iqr]
# df_filtered=df_filtered[df['TMAX']<=Q1+1.5*iqr]
```

```python
def plot_graph(data_array,n_bins,check):

    # binwidth=bin_size
    # Creating histogram 
    fig, axs = plt.subplots(1, 1, figsize =(7, 5)) 
    
    # Add x, y gridlines  
    axs.grid(b = True, color ='grey',  
            linestyle ='-.', linewidth = 0.6,  
            alpha = 0.8)  
    
    # Creating histogram 
    N_val, bins, patches = axs.hist(data_array, bins=n_bins,
                                weights=np.ones_like(data_array) * 1. / len(data_array),cumulative=bool(check)) 
    
    # Setting color 
    color_fraction = ((N_val**(1 / 2)) / N_val.max()) 
    normalized_value = colors.Normalize(color_fraction.min(), color_fraction.max()) 
     
    for frac1, frac2 in zip(color_fraction, patches): 
        color_value = plt.cm.viridis(normalized_value(frac1)) 
        frac2.set_facecolor(color_value) 

    plt.xlabel("Temperature Range") 
    plt.ylabel("Probability") 
    plt.title('Probability Distribution') 
    plt.xlim(left=min(data_array)/2)
    # Show plot 
    plt.show() 
    # print(sum(N_val))
    # print(bins)
```

```python
# shape of df is 7009728 * 29
# print(df.shape)
outlier_airtime=[]
q1,q3,iqr,outlier_airtime=find_outlier(airtime_data_original)
airtime_data=airtime_data[airtime_data_original>=q1-1.5*iqr]
# print(airtime_data.shape)
airtime_data=airtime_data[airtime_data<=q3+1.5*iqr]
# print(airtime_data.shape)
# print()
print("Data at 25th percentile is :" +str(q1))

print("Data at 75th percentile is :" +str(q3))
print("IQR value is :" +str(iqr))
print()
print("Boxplot for Airtime after removing outlier is :")
sns.set_theme(style="whitegrid")  
sns.boxplot(x=airtime_data)
```

```python
print("Probability Distribution for outlier datapoints is :")
plot_graph(outlier_airtime,10,False)
```

```python

airtime_normalized=normalize_data(airtime_data,airtime_data.mean(),airtime_data.std())
print("Mean for Airtime after Normalizing is : " + str(airtime_normalized.mean()))
print("Standard Deviation for Airtime after Normalizing is : " + str(airtime_normalized.std()))
print()
print("Probability distribution of Airtime with 10 bins is :")
plot_graph(airtime_normalized,20, False)
print()
print("Probability distribution of Airtime with 15 bins is :")
plot_graph(airtime_normalized,30, False)
print()
print("Cumulative distribution of Airtime with 10 bins is :")
plot_graph(airtime_normalized,20, True)
print()
print("Cumulative distribution of Airtime with 15 bins is :")
plot_graph(airtime_normalized,30, True)
```
